```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Test dataset from a CSV file
data = pd.read_csv('test_data.csv')
```

```python
data
```

```python
# Separating features and target variable
x = data.drop(columns=['category'])  # Assuming 'species' is the target column
y = data['category']
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=25)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Applying Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(x_train, y_train)
```

```python
# Making predictions on the test set
y_pred = dt_classifier.predict(x_test)
```

```python
y_pred
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
accuracy
```

```python
# Visualizing the decision tree
plt.figure(figsize=(20, 30))
plot_tree(dt_classifier, filled=True, feature_names=x.columns, class_names=np.unique(y), rounded=True)
plt.title('Decision Tree for Test Dataset')
plt.show()
```
