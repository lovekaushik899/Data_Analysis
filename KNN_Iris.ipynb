```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Iris.csv')
```

```python
x=data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y=data['Species']
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=23)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Applying K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
# Making predictions on the test set
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
# Visualizing the dataset with a pair plot
sns.set(style="whitegrid")
# Pairplot to visualize the dataset
sns.pairplot(data, hue='Species', palette="Set2")
plt.show()
```

```python
# Creating scatter plots for different pairs of features
plt.figure(figsize=(15, 10))

# Scatter plot for Sepal Length vs Sepal Width
plt.subplot(2, 2, 1)
sns.scatterplot(data=data, x=data['SepalLengthCm'], y=data['SepalWidthCm'], hue=data['Species'], palette='Set2')
plt.title('Sepal Length vs Sepal Width')

# Scatter plot for Petal Length vs Petal Width
plt.subplot(2, 2, 2)
sns.scatterplot(data=data, x=data['PetalLengthCm'], y=data['PetalWidthCm'], hue=data['Species'], palette='Set2')
plt.title('Petal Length vs Petal Width')

# Scatter plot for Sepal Length vs Petal Length
plt.subplot(2, 2, 3)
sns.scatterplot(data=data, x=data['SepalLengthCm'], y=data['PetalLengthCm'], hue=data['Species'], palette='Set2')
plt.title('Sepal Length vs Petal Length')

# Scatter plot for Sepal Width vs Petal Width
plt.subplot(2, 2, 4)
sns.scatterplot(data=data, x=data['SepalWidthCm'], y=data['PetalWidthCm'], hue=data['Species'], palette='Set2')
plt.title('Sepal Width vs Petal Width')

plt.tight_layout()
plt.show()
```
